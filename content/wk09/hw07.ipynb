Looking at the data, we see that all the assets we have included have maturity dates on Feb. 15, for various years. To simplify calculations, pretend that it is currently Feb. 15, 2019 and we have just retrieved this market data. Therefore, the first asset matures in exactly 1 year, the next in 2 years, and so on.

First, let's remind ourselves the meaning of the discount factor. Taking the first asset given, the one expiring in 1 year, we see that the price of the asset is about $\$105.66$. This price is the sum of discounted future cash flows. Assuming that each of these bonds pays an annual coupon worth 2% of the face value, and all face values are $\$100$, then the price of this first asset was generated by

$$P = (DV_1 \times C) + (DV_1 \times F) = (DV_1 \times (2 + 100)) = 105.65625$$

where $DV_1$ represents the discount factor for a future payment 1 year in the future.

Similarly, for the second bond, the price was found by

$$P = (DV_1 \times C) + (DV_2 \times C) + (DV_2 \times F) = (DV_1 \times 2) + (DV_2 \times (2 + 100)) = 110.1875$$

as the first coupon the bond pays, after 1 year, must be discounted using $DV_1$ but the second coupon and face value payment after 2 years must be discounted by $DV_2$. The fair prices of all of the bonds in this dataset were determined by the market, using the same pattern as above.

You will notice that this pattern is recursive, in a sense, in that to find $DV_i$ you need to know $DV_{i-1}$. Your task for this problem is to find the discount factors for each year into the future, starting at $DV_1$ all the way to $DV_{10}$. You will also notice in the data that there are some jumps in year; for instance, we are missing a bond expiring in 2022. For these cases where we cannot explicitly find a certain $DV_i$, just assume that $DV_i = DV_{i-1}$.
